```python
import matplotlib.pyplot as plt
import json
import numpy as np
import pandas as pd
import requests 
```

```python
def connect_to_api():
    API_KEY = '21PLkuwpdzgX19ne3q7poERRV6u'
    inflow = requests.get('https://api.glassnode.com/v1/metrics/transactions/transfers_volume_to_exchanges_sum',
                           params={'a': 'BTC', 'api_key': API_KEY})
    df_inflow = pd.read_json(inflow.text, convert_dates=['t'])
    balance = requests.get('https://api.glassnode.com/v1/metrics/distribution/balance_exchanges',
                           params={'a': 'BTC', 'api_key': API_KEY})
    df_balance = pd.read_json(balance.text, convert_dates=['t'])
    price = requests.get('https://api.glassnode.com/v1/metrics/market/price_usd_close',
                           params={'a': 'BTC', 'api_key': API_KEY})
    df_price = pd.read_json(price.text, convert_dates=['t'])
    
    data = [df_inflow, df_balance, df_price]
    return data
```

```python
data = connect_to_api()
print(data[0])
```

```python
fig, ax1 = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(14)

ax1.plot(data[0]['t'], data[0]['v'], 'orange')

ax2 = ax1.twinx()
ax2.plot(data[2]['t'], data[2]['v'], 'grey')

ax1.set_xlabel('Date')
ax1.set_ylabel('Exchange Inflow Total')
ax2.set_ylabel('Price')
#ax2.set_yscale('function', functions=(exp))
#ax2.set_yticks([1, 10, 400, 10000, 80000])
```
